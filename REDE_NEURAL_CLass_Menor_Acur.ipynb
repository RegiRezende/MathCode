```python
# ==============================================
#           REDE NEURAL ACURÁCIA 0.954
# ==============================================
# Este modelo é uma rede feedforward com uma camada oculta,
# para classificação multiclasse usando função de ativação softmax
# na camada de saída para transformar os resultados brutos em probabilidades,
# Cross-Entropy (Entropia Cruzada) como função de perda.

# ==============================================
# PARTE 0: IMPORTAÇÕES E CONFIGURAÇÕES
# ==============================================

import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
import pandas as pd
import pickle

# Configurações de visualização
plt.rcParams['figure.figsize'] = (10, 8)
plt.style.use('dark_background')


# ==============================================
# PARTE 1: CLASSE DA REDE NEURAL
# ==============================================

class NnModel:
    """
    Classe que implementa uma rede neural feedforward com uma camada oculta.

    A rede tem a seguinte arquitetura:
    - Camada de entrada: número de neurônios = número de características
    - Camada oculta: número de neurônios configurável (padrão=10)
    - Camada de saída: número de neurônios = número de classes

    Funcionalidades:
    - Forward propagation: calcula as previsões
    - Loss calculation: calcula o erro usando cross-entropy
    - Backpropagation: calcula os gradientes
    - Fit: treina a rede por múltiplas épocas
    """

    def __init__(self, x: np.ndarray, y: np.ndarray, hidden_neurons: int = 10, output_neurons: int = 2):
        """
        Inicializa a rede neural com pesos e biases aleatórios.

        Args:
            x: Dados de entrada (matriz de características)
            y: Rótulos (valores alvo)
            hidden_neurons: Número de neurônios na camada oculta
            output_neurons: Número de neurônios na camada de saída (classes)
        """
        np.random.seed(8)  # Para reprodutibilidade

        # Guardando dados
        self.x = x
        self.y = y

        # Definindo arquitetura
        self.hidden_neurons = hidden_neurons
        self.output_neurons = output_neurons
        self.input_neurons = self.x.shape[1]

        # Inicialização Xavier dos pesos (mantém a variância consistente)
        self.W1 = np.random.randn(self.input_neurons, self.hidden_neurons) / np.sqrt(self.input_neurons)
        self.B1 = np.zeros((1, self.hidden_neurons))
        self.W2 = np.random.randn(self.hidden_neurons, self.output_neurons) / np.sqrt(self.hidden_neurons)
        self.B2 = np.zeros((1, self.output_neurons))

        # Dicionário com todos os parâmetros
        self.model_dict = {'W1': self.W1, 'B1': self.B1, 'W2': self.W2, 'B2': self.B2}

        # Variáveis para armazenar estados intermediários
        self.z1 = 0  # Combinação linear da camada 1 (pré-ativação)
        self.f1 = 0  # Ativação da camada 1 (pós-ativação)

    def forward(self, x: np.ndarray):
        """
        Forward propagation: calcula a saída da rede para uma entrada x.

        Fluxo:
        x → z1 = x·W1 + B1 → f1 = tanh(z1) → z2 = f1·W2 + B2 → softmax(z2)

        Args:
            x: Dados de entrada

        Returns:
            softmax: Probabilidades para cada classe
        """
        # Primeira camada: combinação linear + ativação tanh
        self.z1 = x.dot(self.W1) + self.B1
        self.f1 = np.tanh(self.z1)

        # Segunda camada: combinação linear
        z2 = self.f1.dot(self.W2) + self.B2

        # Softmax: transforma em probabilidades
        exp_values = np.exp(z2)
        softmax = exp_values / np.sum(exp_values, axis=1, keepdims=True)
        return softmax

    def loss(self, softmax):
        """
        Calcula a perda usando cross-entropy.

        Queremos maximizar a probabilidade da classe correta.
        Perda = -log(probabilidade_da_classe_correta)

        Args:
            softmax: Probabilidades previstas

        Returns:
            Perda média (cross-entropy)
        """
        n_samples = self.y.shape[0]
        correct_class_probs = np.zeros(n_samples)

        # Pega a probabilidade da classe correta para cada amostra
        for i in range(n_samples):
            correct_index = self.y[i]
            predicted_prob = softmax[i][correct_index]
            correct_class_probs[i] = predicted_prob

        # Cross-entropy loss
        log_probs = -np.log(correct_class_probs)
        return np.mean(log_probs)

    def backpropagation(self, softmax: np.ndarray, learning_rate: float) -> None:
        """
        Backpropagation: calcula gradientes e atualiza pesos.

        Args:
            softmax: Saída da rede (probabilidades)
            learning_rate: Taxa de aprendizado
        """
        # Gradiente na camada de saída (softmax + cross-entropy)
        delta2 = np.copy(softmax)
        delta2[range(self.x.shape[0]), self.y] -= 1

        # Gradientes dos pesos e biases da camada de saída
        dW2 = (self.f1.T).dot(delta2)
        dB2 = np.sum(delta2, axis=0, keepdims=True)

        # Propagação do erro para camada oculta
        delta1 = delta2.dot(self.W2.T) * (1 - np.power(self.f1, 2))  # derivada da tanh

        # Gradientes dos pesos e biases da camada oculta
        dW1 = (self.x.T).dot(delta1)
        dB1 = np.sum(delta1, axis=0, keepdims=True)

        # Atualização dos pesos (gradient descent)
        self.W1 += -learning_rate * dW1
        self.W2 += -learning_rate * dW2
        self.B1 += -learning_rate * dB1
        self.B2 += -learning_rate * dB2

    def fit(self, epochs: int, lr: float):
        """
        Treina a rede neural.

        Args:
            epochs: Número de épocas de treinamento
            lr: Learning rate

        Returns:
            predictions: Previsões finais para os dados de treinamento
        """
        print(f"\nIniciando treinamento - {epochs} épocas...")

        for epoch in range(epochs):
            # Forward pass
            outputs = self.forward(self.x)

            # Calcula perda
            loss = self.loss(outputs)

            # Backward pass
            self.backpropagation(outputs, lr)

            # Calcula acurácia
            predictions = np.argmax(outputs, axis=1)
            correct = (predictions == self.y).sum()
            accuracy = correct / self.y.shape[0]

            # Mostra progresso a cada 10%
            if int((epoch + 1) % max(1, epochs/10)) == 0:
                print(f'Época: [{epoch+1:4d}/{epochs}]  |  Acurácia: {accuracy:.3f}  |  Perda: {loss:.6f}')

        print("Treinamento concluído!")

        # Retorna previsões finais
        final_outputs = self.forward(self.x)
        return np.argmax(final_outputs, axis=1)


# ==============================================
# PARTE 2: EXPERIMENTO 1 - DATASET MAKE_MOONS
# ==============================================

print("\n" + "="*60)
print("EXPERIMENTO 1: Dataset make_moons (2 classes, 2D)")
print("="*60)

# Gerando dados
x, y = datasets.make_moons(n_samples=500, noise=0.05)

print(f"Forma dos dados: x = {x.shape}, y = {y.shape}")
print("Primeiras 5 amostras:")
print(pd.DataFrame({'x1': x[:5,0], 'x2': x[:5,1], 'y': y[:5]}))

# Verificando distribuição das classes
unique = np.unique(y, return_counts=True)
for label, qt_label in zip(unique[0], unique[1]):
    print(f'Classe {label}: {qt_label} amostras')

# Visualizando dados originais
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(x[:,0], x[:,1], c=y, s=50, alpha=0.5, cmap='cool')
plt.title("Dataset make_moons - Original")
plt.xlabel("x1")
plt.ylabel("x2")

# Configurando e treinando modelo
hidden_neurons = 10
output_neurons = 2
learning_rate = 0.01
epochs = 30

model_moons = NnModel(x, y, hidden_neurons=hidden_neurons, output_neurons=output_neurons)
result_moons = model_moons.fit(epochs, learning_rate)

# Visualizando previsões
plt.subplot(1, 2, 2)
plt.scatter(x[:,0], x[:,1], c=result_moons, s=50, alpha=0.5, cmap='cool')
plt.title(f"Previsões do Modelo - Acurácia: {(result_moons == y).mean():.3f}")
plt.xlabel("x1")
plt.ylabel("x2")
plt.tight_layout()
plt.show()


# ==============================================
# PARTE 3: EXPERIMENTO 2 - DATASET MAKE_BLOBS (2D, 4 CLASSES)
# ==============================================

print("\n" + "="*60)
print("EXPERIMENTO 2: Dataset make_blobs 2D (4 classes)")
print("="*60)

# Gerando dados em 2D com 4 clusters
x_blobs_2d, y_blobs_2d = datasets.make_blobs(
    n_samples=400,
    n_features=2,
    centers=4,
    random_state=10,
    cluster_std=0.9,
    shuffle=True
)

print(f"Forma dos dados: x = {x_blobs_2d.shape}, y = {y_blobs_2d.shape}")
print(f"Classes: {np.unique(y_blobs_2d)}")

# Visualizando dados originais
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(x_blobs_2d[:,0], x_blobs_2d[:,1], c=y_blobs_2d, s=50, alpha=0.7, cmap='cool')
plt.title("Dataset make_blobs 2D - Original")
plt.xlabel("x1")
plt.ylabel("x2")

# Configurando e treinando modelo
hidden_neurons = 50
output_neurons = 4
learning_rate = 0.0015
epochs = 1000

model_blobs_2d = NnModel(x_blobs_2d, y_blobs_2d,
                         hidden_neurons=hidden_neurons,
                         output_neurons=output_neurons)
result_blobs_2d = model_blobs_2d.fit(epochs, learning_rate)

# Visualizando previsões
plt.subplot(1, 2, 2)
plt.scatter(x_blobs_2d[:,0], x_blobs_2d[:,1], c=result_blobs_2d, s=50, alpha=0.7, cmap='cool')
plt.title(f"Previsões do Modelo - Acurácia: {(result_blobs_2d == y_blobs_2d).mean():.3f}")
plt.xlabel("x1")
plt.ylabel("x2")
plt.tight_layout()
plt.show()


# ==============================================
# PARTE 4: EXPERIMENTO 3 - DATASET MAKE_BLOBS (3D, 4 CLASSES)
# ==============================================

print("\n" + "="*60)
print("EXPERIMENTO 3: Dataset make_blobs 3D (4 classes)")
print("="*60)

# Gerando dados em 3D com 4 clusters
x_blobs_3d, y_blobs_3d = datasets.make_blobs(
    n_samples=400,
    n_features=3,
    centers=4,
    random_state=50,
    cluster_std=0.9,
    shuffle=True
)

print(f"Forma dos dados: x = {x_blobs_3d.shape}, y = {y_blobs_3d.shape}")
print("Primeiras 5 amostras (x1, x2, x3, y):")
for i in range(5):
    print(f"  {x_blobs_3d[i]} → classe {y_blobs_3d[i]}")

# Visualizando dados originais em 3D
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
scatter1 = ax1.scatter(x_blobs_3d[:,0], x_blobs_3d[:,1], x_blobs_3d[:,2],
                       c=y_blobs_3d, s=40, alpha=0.7, cmap='cool')
ax1.set_title("Dataset make_blobs 3D - Original")
ax1.set_xlabel("X1")
ax1.set_ylabel("X2")
ax1.set_zlabel("X3")
plt.colorbar(scatter1, ax=ax1, shrink=0.5)

# Configurando e treinando modelo 3D
output_neurons = 4  # Número de classes

model_blobs_3d = NnModel(x_blobs_3d, y_blobs_3d,
                         hidden_neurons=10,
                         output_neurons=output_neurons)
result_blobs_3d = model_blobs_3d.fit(40, 0.001)

# Visualizando previsões em 3D
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
scatter2 = ax2.scatter(x_blobs_3d[:,0], x_blobs_3d[:,1], x_blobs_3d[:,2],
                       c=result_blobs_3d, s=40, alpha=0.7, cmap='cool')
ax2.set_title(f"Previsões do Modelo 3D - Acurácia: {(result_blobs_3d == y_blobs_3d).mean():.3f}")
ax2.set_xlabel("X1")
ax2.set_ylabel("X2")
ax2.set_zlabel("X3")
plt.colorbar(scatter2, ax=ax2, shrink=0.5)

plt.tight_layout()
plt.show()


# ==============================================
# PARTE 5: RESUMO DOS RESULTADOS
# ==============================================

print("\n" + "="*60)
print("RESUMO DOS EXPERIMENTOS")
print("="*60)

print(f"""
Experimento 1 - make_moons (2D, 2 classes):
  • Arquitetura: 2→10→2 neurônios
  • Hiperparâmetros: lr=0.01, épocas=30
  • Acurácia final: {(result_moons == y).mean():.3f}

Experimento 2 - make_blobs 2D (4 classes):
  • Arquitetura: 2→50→4 neurônios
  • Hiperparâmetros: lr=0.0015, épocas=1000
  • Acurácia final: {(result_blobs_2d == y_blobs_2d).mean():.3f}

Experimento 3 - make_blobs 3D (4 classes):
  • Arquitetura: 3→10→4 neurônios
  • Hiperparâmetros: lr=0.001, épocas=40
  • Acurácia final: {(result_blobs_3d == y_blobs_3d).mean():.3f}
""")
```
