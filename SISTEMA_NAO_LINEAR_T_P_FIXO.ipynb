```python
import numpy as np

def ponto_fixo_sistema(G, x0, tolerancia=1e-4, max_iter=200):
    """
    Método do ponto fixo para o sistema não linear:
    x² + y² = 4
    x² - y = 1
    Parâmetros:
    G: função que implementa x = G(x)
    x0: chute inicial (array numpy)
    tolerancia: critério de parada
    max_iter: número máximo de iterações
    """
    x = np.array(x0, dtype=float)

    print(f"Iteração 0: {x}")

    for k in range(max_iter):
        x_novo = G(x)
        erro = np.linalg.norm(x_novo - x)

        print(f"Iteração {k+1}: {x_novo} | Erro: {erro:.2e}")

        if erro < tolerancia:
            print(f"\nConvergência em {k+1} iterações!")
            return x_novo

        x = x_novo

    print("Número máximo de iterações atingido!")
    return x

# Definindo G(x,y) = (sqrt(y+1), sqrt(4 - (sqrt(y+1))^2))
def G(x):
    x_novo = np.sqrt(x[1] + 1)
    y_novo = np.sqrt(4 - x_novo**2)
    return np.array([x_novo, y_novo])

# Executando
if __name__ == "__main__":
    solucao = ponto_fixo_sistema(G, [1.5, 1.3])
    print(f"\nSolução: x = {solucao[0]:.8f}, y = {solucao[1]:.8f}")
```
